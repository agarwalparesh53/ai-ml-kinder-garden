```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Vector Products

### Dot Product

$\hat{a}.\hat{b}$

This tells how much do these point in same direction.

#### How to calculate dot product

$a = \begin{bmatrix} a_1 & a_2 & a_3 \end{bmatrix}$
$b = \begin{bmatrix} b_1 & b_2 & b_3 \end{bmatrix}$

$\hat{a}.\hat{b} = a_1b_1 + a_2b_2 + a_3b_3$


#### Cauchy-Schwarz Inequality

$|\hat{a}.\hat{b}| \: \le ||a||b||$

Equality holds when `a` is a scalar multiple of `b`.

#### Angle between vectors

$\hat{a}.\hat{b} = \|a\| \|b\|\cos\theta$

When dot product is 0, the vectors are orthogonal vectors.

### Cross Product

$\hat{a} \mathbb{X} \hat{b}$ - `Vector which is orthogonal to both a and b`

This tells how much do these point in opposite direction.

#### Angle between vectors

$\hat{a} \mathbb{X} \hat{b} = \|a\| \|b\|\sin\theta$


```python
# Define vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# Compute the cross product
v3 = np.cross(v1, v2)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot vectors
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='${v}_1$', linewidth=2)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label='${v}_2$', linewidth=2)
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='g', label='${v}_1 \\times \\mathbf{v}_2$', linewidth=2)

# Set plot limits
ax.set_xlim([0, 5])
ax.set_ylim([0, 7])
ax.set_zlim([0, 9])

# Set labels
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_title('Vectors and Cross Product')

# Add legend
ax.legend()

# Show grid
ax.grid(True)

# Show plot
plt.show()
```
