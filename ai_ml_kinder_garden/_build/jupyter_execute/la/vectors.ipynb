# Vectors

## Norms and Unit Vectors 

### Norms of a vector 


Norms represent a magnitude of a vector. Norm can be calculated in multiple ways / formats.
Some of the common ones are:

::::{card}

`L1 Norm`
Also known as Manhattan distance. $||\bar{X}||_1 = |X_1| + |X_2| + |X_3| ... + |X_n|$

`L2 Norm`
$||\bar{X}||_2 = \sqrt{X_1^2 + X_2^2 + X_3^2 ... + X_n^2}$

`Infinity Norm`
$||\bar{X}||_\infty = max(|X_1| , |X_2| , |X_3| ... , |X_n|)$

`Unit Vectors`

unit vectors are often represented by `e` which is equivalent to $\frac{\bar{X}}{\sqrt{\sum_{i=1}^N X_i^2}}$
This basically means a vector with magnitude 1 and direction.
::::

## Vector Transformations

```python
V = np.array([1, 4])  
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r', label='V')
plt.quiver(0, 0, -V[0], -V[1], angles='xy', scale_units='xy', scale=1, color='b', label='$-V$')

plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Add labels and grid
plt.xlabel('X')
plt.ylabel('Y')
plt.vlines(1, 0, 4, linestyles="dashed")
plt.hlines(4, 0, 1, linestyles="dashed")
plt.vlines(-1, 0, -4, linestyles="dashed")
plt.hlines(-4, 0, -1, linestyles="dashed")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('V, -V')
plt.legend()
plt.show()
```

```python
V = np.array([1, 4])  
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r', label='V')
plt.quiver(0, 0, V[1], V[0], angles='xy', scale_units='xy', scale=1, color='b', label='$V^T$')

plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Add labels and grid
plt.xlabel('X')
plt.ylabel('Y')
plt.vlines(1, 0, 4, linestyles="dashed")
plt.hlines(4, 0, 1, linestyles="dashed")
plt.hlines(1, 0, 4, linestyles="dashed")
plt.vlines(4, 0, 1, linestyles="dashed")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Transpose Of Vector')
plt.legend()
plt.show()
```

## Basis, Span, Orthogonal and Orthonormal Vectors

::::{card}

`Span`

A span is a set of all vectors which can be represented as a linear combination of a set.
    
`Basis`

Basis represents a set of N length vectors which can represent any vector in N dimension space.

`Orthogonal vectors`

Vectors X, Y are orthogonal if $X^TY = 0$

A N dim space has atmost N orthogonal basis vectors.

`Orthonormal vectors`

Orthogonal vectors which have unit norm.

::::

```python
vector_original = np.random.randint(1, 4, 2)
original_basis = np.array([[0, 1], [1, 0]])
new_basis = np.array([[2, 1], [1, 3]])
vector_new = np.linalg.inv(new_basis) @ vector_original

plt.figure(figsize=(8, 8))

plt.quiver(0, 0, original_basis[0, 0], original_basis[1, 0], angles='xy', scale_units='xy', scale=1, color='r', label='Original Basis')
plt.quiver(0, 0, original_basis[0, 1], original_basis[1, 1], angles='xy', scale_units='xy', scale=1, color='r')

plt.quiver(0, 0, new_basis[0, 0], new_basis[1, 0], angles='xy', scale_units='xy', scale=1, color='b', label='New Basis')
plt.quiver(0, 0, new_basis[0, 1], new_basis[1, 1], angles='xy', scale_units='xy', scale=1, color='b')

plt.quiver(0, 0, vector_original[0], vector_original[1], angles='xy', scale_units='xy', scale=1, color='k', label='Original Vector')
plt.quiver(0, 0, vector_new[0], vector_new[1], angles='xy', scale_units='xy', scale=1, color='y', label='Transformed Vector')

plt.xlim(-4, 4)
plt.ylim(-4, 4)

plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Change of Basis')
plt.legend()
plt.show()
```

## Linearly Independence

### Definition

A set of vectors $(\{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_n\})$ in a vector space is said to be `linearly independent` if the only solution to the equation
$[
c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \cdots + c_n \mathbf{v}_n = \mathbf{0}]$
is $(c_1 = c_2 = \cdots = c_n = \mathbf{0})$.

### Example

Column vectors of $\begin{bmatrix} 1 & 2 & 3 \\ 2 & 4 & 6 \\ 3 & 6 & 7 \end{bmatrix}$ are not linearly indepdenent

Column vectors of $\begin{bmatrix} 1 & 2 & 3 \\ 2 & 4 & 6 \\ 5 & 6 & 7 \end{bmatrix}$ are linearly indepdenent
